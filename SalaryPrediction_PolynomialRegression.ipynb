```python
import pandas as pd
import numpy as np
path = r"C:\Users\Harujah\Desktop\Pantech_Courses\ML\ML_Projects\datasets\salary_polynomial.csv"
df = pd.read_csv(path)
df
```

```python
X = df.iloc[:,:-1].values
X
Y= df.iloc[:,-1].values
Y
```

```python
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression()
modelLR.fit(X,Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,color = "red")
plt.plot(X, modelLR.predict(X))
plt.title("Linear Regression")
plt.xlabel("level")
plt.ylabel("salary")
plt.show()
```

```python
#convert X to Polynomial format (X^n)
# n- degree

from sklearn.preprocessing import PolynomialFeatures
modelPR = PolynomialFeatures(degree=4)
xPoly = modelPR.fit_transform(X)
```

```python
modelPLR = LinearRegression()
modelPLR.fit(xPoly,Y)
```

```python
X=7
salaryPred = modelPLR.predict(modelPR.fit_transform([[X]]))
print('salary of level {0} is {1}'.format(X,salaryPred))
```
